```python
import os
import sys
import django

# Async güvensiz modunu etkinleştir
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

# Django projesinin path'ini ekle
sys.path.append('/home/ocans/DJANGO-AI-AGENT/src')

# Django settings'i tanımla
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'cfehome.settings')

# Django'yu başlat
django.setup()
```

```python
from langgraph.checkpoint.memory import InMemorySaver
from AI.agents import get_document_agent
```

```python
import uuid
config = {"configurable": {"user_id": 2, "thread_id": uuid.uuid4()}}
checkpointer = InMemorySaver()
agent = get_document_agent(model="gemini-2.5-flash", checkpointer=checkpointer)
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "Give me a summary of the movie Mission Impossible?"}
        ]
    },
    config
)
response["messages"][-1].content
```
