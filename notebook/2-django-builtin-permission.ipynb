```python
from django.contrib.auth import get_user_model

User = get_user_model()

# list -> get 
User.objects.all()
```

```python
superuser = User.objects.filter(is_superuser=True).first()
superuser
```

```python
from django.contrib.auth.models import Permission
perms_qs = Permission.objects.all()
for perm in perms_qs:
    print(perm.codename, "-", perm.name)
```

```python
staff = User.objects.filter(is_superuser=False, is_staff=True).first()
staff
```

```python
perms_qs = Permission.objects.filter(codename__icontains="VIEW")
for obj in perms_qs:
    print(obj.codename)
    staff.user_permissions.remove(obj)
```

```python
docs_qs = Permission.objects.filter(codename__endswith="document")
docs_qs
#staff.user_permissions.set(perms_qs)
```

```python
new_perms = docs_qs | perms_qs
new_perms
```

```python
new_perms = new_perms.exclude(codename__icontains='delete')
staff.user_permissions.set(new_perms)
```
